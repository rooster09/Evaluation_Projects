```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
coln = list(maindt.columns)
colcount = len(maindt.columns)
x=0
for col in maindt.columns:
    if x < colcount:
        sns.distplot(maindt[coln[x]])
        x=x+1
        plt.show()
```

Data seems to be pretty much properly distributed not much skewness detected. Next is to examine outliers using box plot.

```python
plt.figure(figsize=(20,50))
colcount = len(maindt.columns)
x=1
for col in maindt:
    if x <= colcount:
        ax= plt.subplot(10,3,x)
        sns.boxplot(maindt[col],orient='v')
        plt.xlabel(col,fontsize=10)
    x=x+1
plt.show() 
```

```python
Q1 = maindt['E'].quantile(0.25)
Q3 = maindt['E'].quantile(0.75)
IQR = Q3 - Q1    #IQR is interquartile range. 

filter = (maindt['E'] >= Q1 - 1.5 * IQR) & (maindt['E'] <= Q3 + 1.5 *IQR)
maindt.loc[filter]  
```

```python
plt.figure(figsize=(20,50))
colcount = len(maindt.columns)
x=1
for col in maindt:
    if x <= colcount:
        ax= plt.subplot(10,3,x)
        sns.boxplot(maindt[col],orient='v')
        plt.xlabel(col,fontsize=10)
    x=x+1
plt.show() 
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
y=maindt['W']
x=maindt.drop('W',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
rf=RandomForestRegressor()
lr= LinearRegression()
dr=DecisionTreeRegressor()
sv=SVR()
```

For Linear Regression.

```python
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    print('R2 score = ', r2_score(y_test,pred))
    print('Mean square error =',mean_squared_error(y_test,pred))
    print('Absolute error =',mean_absolute_error(y_test,pred))
    print('RMS value =',np.sqrt(mean_squared_error(y_test,pred)))
```

For Decision Tree Regressor

```python
    dr.fit(x_train,y_train)
    pred=dr.predict(x_test)
    print('R2 score = ', r2_score(y_test,pred))
    print('Mean square error =',mean_squared_error(y_test,pred))
    print('Absolute error =',mean_absolute_error(y_test,pred))
    print('RMS value =',np.sqrt(mean_squared_error(y_test,pred)))
```

For SVR

```python
    sv.fit(x_train,y_train)
    pred=sv.predict(x_test)
    print('R2 score = ', r2_score(y_test,pred))
    print('Mean square error =',mean_squared_error(y_test,pred))
    print('Absolute error =',mean_absolute_error(y_test,pred))
    print('RMS value =',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
    rf.fit(x_train,y_train)
    pred=rf.predict(x_test)
    print('R2 score = ', r2_score(y_test,pred))
    print('Mean square error =',mean_squared_error(y_test,pred))
    print('Absolute error =',mean_absolute_error(y_test,pred))
    print('RMS value =',np.sqrt(mean_squared_error(y_test,pred)))
```
