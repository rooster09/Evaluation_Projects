Converting text data to numerical with label encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
maindt['Gender'] = encoder.fit_transform(maindt['Gender'])
maindt['JobRole'] = encoder.fit_transform(maindt['JobRole'])
maindt['MaritalStatus'] = encoder.fit_transform(maindt['MaritalStatus'])
maindt['OverTime'] = encoder.fit_transform(maindt['OverTime'])
maindt['BusinessTravel'] = encoder.fit_transform(maindt['BusinessTravel'])
maindt['Department'] = encoder.fit_transform(maindt['Department'])
maindt['EducationField'] = encoder.fit_transform(maindt['EducationField'])
maindt['Attrition'] = encoder.fit_transform(maindt['Attrition'])
```

Finding the correlation between the features

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
sns.heatmap(maindt.corr(), color = "k", annot=True  )
```

```python
sns.countplot(maindt['Attrition']);
```

Class imbalance issue

```python
x=maindt.drop("Attrition",axis=1)
y = maindt['Attrition']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
obj = SMOTE()
x, y = obj.fit_resample(x,y)
```

```python
x.describe()
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

Transforming values to standard values uniformly

```python
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x))
```

```python
x.head()
```

Building Models

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 27)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)

dtcscore = dtc.score(x_test,y_test)
print("Model Score", dtcscore )
print("Confusion Matrix",confusion_matrix(y_test, dtc.predict(x_test)))
dtccvscore = cross_val_score(dtc,x,y,cv=5)
print("Cross Valudition", np.mean(dtccvscore) )
dtcpred = dtc.predict(x_test)
print("Classification Report", classification_report(y_test,dtcpred))
```

Decision tree Classifier Difference = 1%

KNN Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)

knnscore = knn.score(x_test,y_test)
print("Model Score", knnscore )
print("Confusion Matrix",confusion_matrix(y_test, knn.predict(x_test)))
knncvscore = cross_val_score(knn,x,y,cv=5)
print("Cross Valudition", np.mean(knncvscore) )
knnpred = knn.predict(x_test)
print("Classification Report", classification_report(y_test,knnpred))
```

KNN Neighbours difference 1%

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)

svcscore = svc.score(x_test,y_test)
print("Model Score", svcscore )
print("Confusion Matrix",confusion_matrix(y_test, svc.predict(x_test)))
svccvscore = cross_val_score(svc,x,y,cv=5)
print("Cross Valudition", np.mean(svccvscore) )
svcpred = svc.predict(x_test)
print("Classification Report", classification_report(y_test,svcpred))
```

SVC model difference 2% with 90% accuracy

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfcmodel = RandomForestClassifier()
rfcmodel.fit(x_train,y_train)

rfcscore = rfcmodel.score(x_test,y_test)
print("Model Score", dtcscore )
print("Confusion Matrix",confusion_matrix(y_test, svc.predict(x_test)))
rfccvscore = cross_val_score(rfcmodel,x,y,cv=5)
print("Cross Valudition", np.mean(rfccvscore) )
rfcpred = rfcmodel.predict(x_test)
print("Cross Valudition", classification_report(y_test,rfcpred))
```

Random Forest giving 90% accuracy with less than 1% difference.

Logistic Regression as binary classification is there.

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(x_train,y_train)

lgscore = lg.score(x_test,y_test)
print("Model Score", lgscore )
print("Confusion Matrix",confusion_matrix(y_test, lg.predict(x_test)))
lgcvscore = cross_val_score(lg,x,y,cv=5)
print("Cross Valudition", np.mean(lgcvscore) )
lgpred = lg.predict(x_test)
print("Classification Report", classification_report(y_test,lgpred))
```

Logistic Regression with difference of 1 %
